## Introduction

The Rijksmuseum in Amsterdam has been widely recognised as a pioneering institution in making cultural heritage openly accessible to researchers and the wider public (Terras, 2015). The digital transformation of the Dutch museums can be traced back to 1969 when the government significantly subsidised the museum sector, favouring the first wave of digitisation (Navarrete, 2014). Throughout the decades, the Rijksmuseum has not only integrated museum data into its computer system but also made more than 823,077 artworks, 348,151 library items, and 196,648 visitor stories publicly available online (Rijksmuseum, 2025). The milestones of openness can be allocated on the timeline, mapping the release of a substantial portions of its resources into the public domain under a Creative Commons Zero (CC0) license since 2011; the publication of its digital collections as Linked Open Data (Dijkshoorn et al., 2018); the promotion of use and reuse of its digital collections in high resolution through initiatives such as the Rijksawards and the Rijksstudio (Volkers, 2017); and the development of tools for data search and exploration (Mensink & van Gemert, 2014), alongside other projects. In fact, having these noticeable milestones described and accounted for, the historical perspective of the development of the museumâs digital transformation in openness remains understudied, leaving significant gaps in understanding how it emerged, was implemented, and developed within this cultural heritage institution.

Do we have sufficient source base to analyse the shift in openness or understand the priorities in digital publishing of information on the web by the museum? Can we apply data-driven methodologies and inspect the data uploaded on the website over the past years? The opportunities to answer these questions depend mostly on the materials that were preserved by the web archives. To trace historical developments and investigate the museumâs web presence, analysing web-archived materials is essential. Web archives preserve unique snapshots of the museumâs online presence, serving as a historical record, invaluable for such study (BrÃ¼gger, 2017). These repositories, particularly the Internet Archive and the Royal Library of the Netherlands (the Koninklijke Bibliotheek, KB), which is responsible for preserving the Dutch web, crawl websites to archive them for the future. At the same time, the web archives present unique challenges for historians: they are partial, sometimes inconsistent, and shaped by crawler technologies, and largely affected by the selection policies that determine what survives for future analysis and what remains omitted (REFS). As a fundamental practice in historical research, source criticism is a necessary step for evaluating this source base and, to determine what kinds of research questions we, as researchers, can raise, the questions of accuracy and completeness of web archive data must be addressed (Weber, 2018; REFS).

By applying source criticism to these digital repositories, this study systematically observes datasets that document the Rijksmuseumâs web presence in historical depth, revealing both the possibilities and limitations of using materials from the Internet Archive and the KB as historical sources. It analyses these web-archived collections to determine the coverage of preserved materials over time, identify missing information, evaluate gaps, and assess overlaps that could support building a more comprehensive combined dataset. The study also examines how effectively these two major web archiving initiatives have captured the Rijksmuseumâs online presence and identifies temporal blind spots within the archival records.
