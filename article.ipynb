## Introduction

The Rijksmuseum in Amsterdam has been widely recognised as a pioneering institution in making cultural heritage openly accessible to researchers and the wider public (Terras, 2015). This initiative provided an impetus and became a starting point for the Open GLAM movement, when museums across the globe started making their collections available using open licenses (McCarthy, 2019; Wallace, 2020). This research was largely motivated by the question about the historical development of openness of the Rijksmuseum.

The digital transformation of the Dutch museums can be traced back to 1969 when the government significantly subsidised the museum sector, favouring the first wave of digitisation (Navarrete, 2014). This step and underlying initiatives contributed to preparing the digital infrastructure, which subsequently expanded. The milestones of openness can be allocated on the timeline, mapping the release of a substantial portions of its resources into the public domain under a Creative Commons Zero (CC0) license since 2011; the publication of its digital collections as Linked Open Data (Dijkshoorn et al., 2018); the promotion of use and reuse of its digital collections in high resolution through initiatives such as the Rijksawards and the Rijksstudio (Volkers, 2017); and the development of tools for data search and exploration (Mensink & van Gemert, 2014), alongside other projects. In the figure below, you may see these milestones indicated on the timeline. Today, the Rijksmuseum made 823,077 artworks, 348,151 library items, and 196,648 visitor stories publicly available online (Rijksmuseum, 2025).

Addressing these and other relevant questions would contribute significantly to our understanding of the historical development of open practices.

However, to be able to address these questions, it is necessary to evaluate the available records. Therefore, the main goal of this paper is to assess the availability and evaluate the resources for the historical development of the Rijksmuseum on the web. The museumâs web presence is essential for consideration because the website documents the museumâs various activities and serves as the primary site for distributing information in various formats. 

## Research Goals and Methodology

Research on museums on the web using web archives is a highly interdisciplinary field combining museology, history, information science and web archiving perspectives (de Wild & Povroznik, 2024). Exploring the museums' web presence from a historical perspective is possible mainly through web archives because these platforms are dedicated to the preservation and representation of the past web. Web archives preserve unique snapshots of the museumâs online presence, serving as a historical record invaluable for such study. These repositories, particularly the Internet Archive and the Royal Library of the Netherlands (the Koninklijke Bibliotheek, KB), which is responsible for preserving the Dutch web, crawl websites to archive them for the future.

Scholarship on web history and the use of web archives, particularly the Internet Archive and national web archiving initiatives, has highlighted how web archives can be treated as sources and infrastructures for historical research (Balbi, 2025; BrÃ¼gger, 2018 and 2017; Weber, 2018) and has provided methodological foundations for studying the web (Skov & Svarre, 2024; Povroznik, 2025). Scholars and web archivists also stressed limitations of the web archives for their scholarly exploration (Ben-David & Huurdeman, 2014). These discussions are especially relevant for examining websites, as their digital prints are shaped by fragmented preservation, selection policies of web archives, their technical capacities, priorities and approaches to preservation in general and retrieval issues (BrÃ¼gger & Laursen, 2020; Winters & Prescott, 2019). Such factors determine which portions of the webâs past are captured and what will consequently be accessible to other generations and to scholarship. Therefore, the web study should necessarily integrate source criticism in the work. As a fundamental practice in historical research, source criticism is a necessary step for evaluating this source base and, to determine what kinds of research questions we, as researchers, can raise, the questions of accuracy and completeness of web archive data must be addressed (Weber, 2018; REFS).

Web archives serve as repositories of unique sources documenting the history of museum information resources and the cultural segment of the World Wide Web (Povroznik, 2020). They enable researchers to analyse the structure and content of museum websites, track various museum activities, and observe changes in museums' digital practices over time. A wide range of methodological approaches can be applied, from close reading to distant reading and their mixture (Povroznik, 2025; Skov & Svarre, 2024).

This paper contributes to the methodology of web history by proposing frameworks for evaluating the reliability and completeness of web archives as historical sources. Its findings benefit researchers studying the web presence of the Rijksmuseum or other institutions and help develop best practices for using web-archived resources in historical analysis. For the museum, the study offers structured documentation of its institutional web records over time, supporting strategic reflection on future steps in preserving its web heritage. For web archives, particularly the KB, the analysis provides insights into coverage patterns and preservation challenges that can guide future collection development.

The analysis is limited to the period from 1999 to June 2025 due to the availability of data and limits with the time when the data were collected and analysed. The first preserved snapshot of the Rijksmuseumâs website in the Internet Archive dates back to 28 November 1999, marking the beginning of its preservation there, while the Web collection at the KB, National Library of the Netherlands (Web archiving, 2025) dates back to capturing the site since 2010. This study systematically observes datasets that document the Rijksmuseumâs web presence in historical depth, comparing materials from the Internet Archive and the KB. It analyses these web-archived collections to determine the coverage of preserved materials over time, to identify overlapping and missing information in both datasets, to evaluate gaps, and to assess possibilities for building a more comprehensive combined dataset. The study also examines how effectively these two major web archiving initiatives have captured the Rijksmuseumâs online presence and identifies temporal blind spots within the archival records.

The overarching goal of analysis of the materials about the Rijksmuseum web presence available for analysis on the Internet Archive and the KB Web Collection has been addressed by accomplishing the following tasks:
1) Analysis of the temporal distribution of the data in the datasets;
2) Research the reflection of the museumâs object collection in the datasets;
3) Comparative study of the datasets to identify unique and overlapping materials;
4) Checking the outputs by aligning the web-archived data with the current museumâs data.

These questions were addressed primarily through URL analysis. URLs provide information about content that constituted the Rijksmuseumâs website and its format. The URL string may disclose some of the sections that the website carries. Therefore, it is possible to conduct content analysis of this textual information, identify some sections (such as collection,  and provide insights into the structure of the website and also into the distribution of information across language versions.

## Datasets: Building and Overview

To access data from the Internet Archive, the Wayback Machine API was used (Wayback Machine APIs, 2025). The KB Web Collection was explored based on the Solwayback instance available at the KB Datalab. The datasets were produced by applying common practices in filtering, cleaning, and processing the data. The resulting export file contains metadata for each capture including the original URL, timestamp, status code, MIME type, content digest, and some technical fields. Datasets were shaped by including only the URLs filtered by status 200. The necessary condition was to receive successfully processed data, which is indicated in the metadata with status 200. The other materials were not taken into account as having technical issues or failures during crawling. The function âdigestâ has been applied to filter data and work with unique materials and avoid noise. Digest in metadata reflects the hash value, generated automatically based on the captured content (the bytes), and it is used to identify identical or near-identical versions of the web content.

A series of cleaning steps was implemented to make the data analytically usable. For that reason, the duplicates were removed on the basis of the âdigestâ value. Deduplication was performed separately for each year. Also, the URL strings were filtered by excluding URLs that contained query symbols (such as â?â and â&â), fragments (#), temporary and script-generated parameters. On the next step, the unique URLs were filtered, and as a result, only one unique URL could appear in the data per year, but could repeat through different years.

By applying these steps to data from the Internet Archive and the KB Web Collection, two datasets were obtained. They were used for the subsequent filtering and analysis.

To compare datasets with the original numbers of uploads to the Rijksmuseumâs websites, the statistics were obtained from the Wayback Machine. The Rijksstudio exposed the number of items in the database for both the general pool of artworks and those published in the Rijksstudio. For example, on the snapshot created on 23 October 2018, these numbers are available: 647.965 artworks and 437.977 Rijksstudio's (Rijksmuseum, 2018). Based on the available data from the last successfully preserved snapshot, the diagram was built to be able to trace the progress in uploading objects to the digital collection online. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# If the file is in the same folder as the notebook:
csv_path = "./media/rijksstudio_yearend_counts.csv"

# Load data
df = pd.read_csv(csv_path)

# Quick sanity check
display(df.head())

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df["year"], df["total_artworks"],
         marker="o", label="Total artworks")
plt.plot(df["year"], df["rijksstudios"],
         marker="o", label="Rijksstudio's")

plt.title("Rijksmuseum â Year-end counts of artworks and Rijksstudio's")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

The line chart shows the Rijksmuseumâs total digitised artworks (blue line) and the works available on the Rijksstudio (orange line) from 2013 to 2024. The website indicated the total amount of works available through the database since 2013 according to the records on the Wayback Machine. Both curves display a consistent and smooth growth trajectories, which go closer to each other approaching 2025, which means uploading almost all the digitised the artworks for their reuse through the Rijkstudio platform. Rijksstudio entries steadily grew from about 110,000 to nearly 780,000 in 2024. Having these numbers at stake, it is possible to assess roughly the portion of the museumâs web preserved by the Internet Archive and the KB.

The diagram visualises how the KB has preserved the domain rijksmuseum.nl between 2009 and 2025, based on the size of the collection in kilobytes (blue line), the number of pages (red line) and ingoing links (green line). There are two major preservation waves. The first peak occurs between 2010 and 2014, and by the positioning of the blue curve, it is possible to guess that most of the volume was taken by the images.
