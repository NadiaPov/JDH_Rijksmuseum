## Introduction

The Rijksmuseum in Amsterdam has been widely recognised as a pioneering institution in making cultural heritage openly accessible to researchers and the wider public (Terras, 2015). This initiative provided an impetus and became a starting point for the Open GLAM movement, when museums across the globe started making their collections available using open licenses (McCarthy, 2019; Wallace, 2020). This research was largely motivated by the question about the historical development of openness of the Rijksmuseum.

The digital transformation of the Dutch museums can be traced back to 1969 when the government significantly subsidised the museum sector, favouring the first wave of digitisation (Navarrete, 2014). This step and underlying initiatives contributed to preparing the digital infrastructure, which subsequently expanded. The milestones of openness can be allocated on the timeline, mapping the release of a substantial portions of its resources into the public domain under a Creative Commons Zero (CC0) license since 2011; the publication of its digital collections as Linked Open Data (Dijkshoorn et al., 2018); the promotion of use and reuse of its digital collections in high resolution through initiatives such as the Rijksawards and the Rijksstudio (Volkers, 2017); and the development of tools for data search and exploration (Mensink & van Gemert, 2014), alongside other projects. In the figure below, you may see these milestones indicated on the timeline. Today, the Rijksmuseum made 823,077 artworks, 348,151 library items, and 196,648 visitor stories publicly available online (Rijksmuseum, 2025).

At the same time, do we have an opportunity to explore in more detail how openness was implemented, including how representation of collections grew and changed through the years (if changed at all), which paintings were prioritised for publishing and why, how the structure of metadata changed, and how content developed and expanded? Addressing these and other relevant questions would contribute significantly to our understanding of the historical development of open practices.

However, to be able to address these questions, it is necessary to evaluate the available records. There are three cohorts of resources: 1) datasets and other resources published on the museumâs website; 2) relevant records published on the external platforms (such as Wikidata (2025)); 3) snapshots on the web archives (Internet Archive (2025) and the web archiving collections on the Royal Library of the Netherlands (KB, 2025).

Therefore, the main goal of this paper is to assess the availability and evaluate the resources for the historical development of the Rijksmuseum on the web. The museumâs web presence is essential for consideration because the website documents the museumâs various activities and serves as the primary site for distributing information in various formats. 

## Research Goals and Methodology

Research on museums on the web using web archives is a highly interdisciplinary field combining museology, history, information science and web archiving perspectives (de Wild & Povroznik, 2024). Exploring the museums' web presence from a historical perspective is possible mainly through web archives because these platforms are dedicated to the preservation and representation of the past web. Web archives preserve unique snapshots of the museumâs online presence, serving as a historical record invaluable for such study. These repositories, particularly the Internet Archive and the Royal Library of the Netherlands (the Koninklijke Bibliotheek, KB), which is responsible for preserving the Dutch web, crawl websites to archive them for the future.

Scholarship on web history and the use of web archives, particularly the Internet Archive and national web archiving initiatives, has highlighted how web archives can be treated as sources and infrastructures for historical research (Balbi, 2025; BrÃ¼gger, 2018 and 2017; Weber, 2018) and has provided methodological foundations for studying the web (Skov & Svarre, 2024; Povroznik, 2025). Scholars and web archivists also stressed limitations of the web archives for their scholarly exploration (Ben-David & Huurdeman, 2014). These discussions are especially relevant for examining websites, as their digital prints are shaped by fragmented preservation, selection policies of web archives, their technical capacities, priorities and approaches to preservation in general and retrieval issues (BrÃ¼gger & Laursen, 2020; Winters & Prescott, 2019). Such factors determine which portions of the webâs past are captured and what will consequently be accessible to other generations and to scholarship. Therefore, the web study should necessarily integrate source criticism in the work. As a fundamental practice in historical research, source criticism is a necessary step for evaluating this source base and, to determine what kinds of research questions we, as researchers, can raise, the questions of accuracy and completeness of web archive data must be addressed (Weber, 2018; REFS).

Web archives serve as repositories of unique sources documenting the history of museum information resources and the cultural segment of the World Wide Web (Povroznik, 2020). They enable researchers to analyse the structure and content of museum websites, track various museum activities, and observe changes in museums' digital practices over time. A wide range of methodological approaches can be applied, from close reading to distant reading and their mixture (Povroznik, 2025; Skov & Svarre, 2024).

This paper contributes to the methodology of web history by proposing frameworks for evaluating the reliability and completeness of web archives as historical sources. Its findings benefit researchers studying the web presence of the Rijksmuseum or other institutions and help develop best practices for using web-archived resources in historical analysis. For the museum, the study offers structured documentation of its institutional web records over time, supporting strategic reflection on future steps in preserving its web heritage. For web archives, particularly the KB, the analysis provides insights into coverage patterns and preservation challenges that can guide future collection development.

The analysis is limited to the period from 1999 to June 2025 due to the availability of data and limits with the time when the data were collected and analysed. The first preserved snapshot of the Rijksmuseumâs website in the Internet Archive dates back to 28 November 1999, marking the beginning of its preservation there, while the Web collection at the KB, National Library of the Netherlands (Web archiving, 2025) dates back to capturing the site since 2010. This study systematically observes datasets that document the Rijksmuseumâs web presence in historical depth, comparing materials from the Internet Archive and the KB. It analyses these web-archived collections to determine the coverage of preserved materials over time, to identify overlapping and missing information in both datasets, to evaluate gaps, and to assess possibilities for building a more comprehensive combined dataset. The study also examines how effectively these two major web archiving initiatives have captured the Rijksmuseumâs online presence and identifies temporal blind spots within the archival records.

The overarching goal of analysis of the materials about the Rijksmuseum web presence available for analysis on the Internet Archive and the KB Web Collection has been addressed by accomplishing the following tasks:
1) Analysis of the temporal distribution of the data in the datasets;
2) Research the reflection of the museumâs object collection in the datasets;
3) Comparative study of the datasets to identify unique and overlapping materials;
4) Checking the outputs by aligning the web-archived data with the current museumâs data.

These questions were addressed primarily through URL analysis. URLs provide information about content that constituted the Rijksmuseumâs website and its format. The URL string may disclose some of the sections that the website carries. Therefore, it is possible to conduct content analysis of this textual information, identify some sections (such as collection,  and provide insights into the structure of the website and also into the distribution of information across language versions.

## Datasets: Building and Overview

To access data from the Internet Archive, the Wayback Machine API was used (Wayback Machine APIs, 2025). The KB Web Collection was explored based on the Solwayback instance available at the KB Datalab. The datasets were produced by applying common practices in filtering, cleaning, and processing the data. The resulting export file contains metadata for each capture including the original URL, timestamp, status code, MIME type, content digest, and some technical fields. Datasets were shaped by including only the URLs filtered by status 200. The necessary condition was to receive successfully processed data, which is indicated in the metadata with status 200. The other materials were not taken into account as having technical issues or failures during crawling. The function âdigestâ has been applied to filter data and work with unique materials and avoid noise. Digest in metadata reflects the hash value, generated automatically based on the captured content (the bytes), and it is used to identify identical or near-identical versions of the web content.

A series of cleaning steps was implemented to make the data analytically usable. For that reason, the duplicates were removed on the basis of the âdigestâ value. Deduplication was performed separately for each year. Also, the URL strings were filtered by excluding URLs that contained query symbols (such as â?â and â&â), fragments (#), temporary and script-generated parameters. On the next step, the unique URLs were filtered, and as a result, only one unique URL could appear in the data per year, but could repeat through different years.

By applying these steps to data from the Internet Archive and the KB Web Collection, two datasets were obtained. They were used for the subsequent filtering and analysis.

To compare datasets with the original numbers of uploads to the Rijksmuseumâs websites, the statistics were obtained from the Wayback Machine. The Rijksstudio exposed the number of items in the database for both the general pool of artworks and those published in the Rijksstudio. For example, on the snapshot created on 23 October 2018, these numbers are available: 647.965 artworks and 437.977 Rijksstudio's (Rijksmuseum, 2018). Based on the available data from the last successfully preserved snapshot, the diagram was built to be able to trace the progress in uploading objects to the digital collection online. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

csv_path = "./media/rijksstudio_yearend_counts.csv"
df = pd.read_csv(csv_path)

# ---- Remove 2025 (empty) ----
df = df[df["year"] != 2025].copy()

# ---- Plot (all remaining years) ----
years = df["year"].astype(int)

plt.figure(figsize=(11, 6))

plt.plot(years, df["total_artworks"],
         marker="o", label="Total artworks")
plt.plot(years, df["rijksstudios"],
         marker="o", label="Rijksstudio's")

plt.title("Rijksmuseum â Year-end counts of artworks online and Rijksstudio's entries")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True, linestyle="--", linewidth=0.5)
plt.legend()
plt.xticks(years, rotation=45)
plt.tight_layout()
plt.show()

# ---------- Clean formatting ----------
def fmt_int(x):
    if pd.isna(x):
        return ""
    return f"{int(x)}"

def small_table(df_part):
    tdf = df_part[["year", "total_artworks", "rijksstudios"]].copy()
    return (
        tdf.style
           .format({
               "year": lambda x: f"{int(x)}",
               "total_artworks": fmt_int,
               "rijksstudios": fmt_int
           })
           .set_properties(**{
               "text-align": "center",
               "font-size": "11pt",
               "border": "1px solid #cccccc",
               "padding": "4px"
           })
           .set_table_styles([
               dict(selector="th", props=[
                   ("font-size", "11pt"),
                   ("text-align", "center"),
                   ("border", "1px solid #cccccc"),
                   ("padding", "4px"),
               ])
           ])
           .hide(axis="index")
           .to_html()
    )

# ---- Four fragments ----
t1 = small_table(df[(df["year"] >= 2013) & (df["year"] <= 2015)])
t2 = small_table(df[(df["year"] >= 2016) & (df["year"] <= 2018)])
t3 = small_table(df[(df["year"] >= 2019) & (df["year"] <= 2021)])
t4 = small_table(df[(df["year"] >= 2021) & (df["year"] <= 2024)])

# ---- Show all four tables horizontally ----
html = f"""
<div style="display: flex; gap: 20px;">
    <div>{t1}</div>
    <div>{t2}</div>
    <div>{t3}</div>
    <div>{t4}</div>
</div>
"""

display(HTML(html))
```

The website indicated the total amount of works available through the database since 2013 according to the records on the Wayback Machine. Both curves display a consistent and smooth growth trajectories, which go closer to each other approaching 2025, which means uploading almost all the digitised the artworks for their reuse through the Rijkstudio platform. ### Comparing Datasets

Displaying datasets together on the timeline highlights the uneven distribution of data and different web preservation efforts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import IntText, HBox, VBox, HTML, Output
from IPython.display import display

# === Load Data ===
df = pd.read_csv('./media/url_count_per_year_comparison.csv')
df['Year'] = df['Year'].astype(int)

year_min = df['Year'].min()
year_max = df['Year'].max()


instruction = HTML("<b>Please specify the period you would like to display (1999â2025):</b>")

start_year_widget = IntText(
    value=year_min,
    description='Start year:',
    layout={'width': '200px'}
)

end_year_widget = IntText(
    value=year_max,
    description='End year:',
    layout={'width': '200px'}
)


plot_output = Output()

def update_plot(change=None):
    zoom_start = start_year_widget.value
    zoom_end = end_year_widget.value

    
    if zoom_start > zoom_end:
        zoom_start, zoom_end = zoom_end, zoom_start

    df_zoom = df[(df['Year'] >= zoom_start) & (df['Year'] <= zoom_end)].copy()

    with plot_output:
        plot_output.clear_output(wait=True)

        if df_zoom.empty:
            print(f"No data available for the period {zoom_start}â{zoom_end}.")
            return

        df_zoom['Year'] = df_zoom['Year'].astype(str)

        x = np.arange(len(df_zoom['Year']))
        width = 0.35

        plt.figure(figsize=(16, 6))
        plt.bar(x - width/2, df_zoom['IA'], width, label='IA', color='#1f77b4')   # blue
        plt.bar(x + width/2, df_zoom['KB'], width, label='KB', color='#ff7f0e')   # orange

        plt.title(f'Total URLs per Year ({zoom_start}â{zoom_end})', fontsize=14)
        plt.xlabel('Year')
        plt.ylabel('Number of URLs')
        plt.xticks(x, df_zoom['Year'], rotation=45)
        plt.legend()
        plt.tight_layout()
        plt.show()


start_year_widget.observe(update_plot, 'value')
end_year_widget.observe(update_plot, 'value')


update_plot()


ui = VBox([
    plot_output,
    instruction,
    HBox([start_year_widget, end_year_widget])
])

display(ui)
```

There are some periods in the history of the Rijksmuseum web presence that are covered only by the Internet Archive before 2010 and between 2015 and 2019. At the same time, the KB contributed more significantly to web preservation than the Internet Archive during 2010-2012, and collected more in 2013-2014. These observations show that using a single source of information is not able to cover all periods of the Rijksmuseumâs web presence and that these datasets have the potential to be compiled to get a more comprehensive dataset.

### URLs Analysis

#### Formats

URL strings enable us to identify file formats by analysing their endings. 
The identified file formats can be divided into several functional groups:
1) Core web document formats such as HTML, PHP, and JSP. The HTML is the most important for tracing because it carries content information, so crucial for research. Index files were also included in this group.
2) Image formats (GIF, JPG, PNG, SVG). They provide a record of captured images of paintings. As well as additional images that reflect the design of the website, such as icons, buttons and other fragments of graphics.  
3) Audio and video formats (MP3, MOV).
4) Document formats such as PDF and TXT reflect downloadable materials. 
5) Front-end presentation and scripting formats (CSS, JS, LESS) are visible mainly until 2014, after which many stylesheets and scripts likely moved to external CDNs, resulting in fewer captures.  
6) Other formats that reflect specific components of the website to support its structure and design. 


The most relevant for considering the cultural content are the files containing text (which is represented mainly in the primary HTML format) and images. Comparing their proportions on the datasets will show the focus of preservation efforts over time and will help to assess which expressive aspects of the Rijksmuseum web presence are most visible in these web archival collections.

The two diagrams below display the proportions of HTML and IMAGE formats captured each year, separately for the Internet Archive (IA) and the KB Web Collection (KB). Each bar represents 100% of only HTML and IMAGE types of formats for a given year, but not all the formats in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Paths to your aggregated 2-row tables (HTML / IMAGE) ---
ia_path = "./media/IA_Rijksmuseum_file_formats_per_year_aggregated.csv"
kb_path = "./media/KB_URLs_200statuscode_full_cleaned_Formats_aggregated.csv"

# --- Load data and set index = format ---
ia = pd.read_csv(ia_path).set_index("format")
kb = pd.read_csv(kb_path).set_index("format")

# Make sure all non-format columns are numeric
ia = ia.apply(pd.to_numeric, errors="ignore")
kb = kb.apply(pd.to_numeric, errors="ignore")

years_full = list(range(1999, 2026))  # 1999â2025

def prep_props(df):
    """Return two lists (html_props, image_props) aligned with years_full."""
    html_props = []
    image_props = []
    for y in years_full:
        col = str(y)
        if col in df.columns:
            html_val = df.loc["HTML", col] if "HTML" in df.index else 0
            image_val = df.loc["IMAGE", col] if "IMAGE" in df.index else 0
            total = html_val + image_val
            if total > 0:
                html_props.append(html_val / total)
                image_props.append(image_val / total)
            else:
                html_props.append(np.nan)   # no bar
                image_props.append(np.nan)
        else:
            html_props.append(np.nan)       # no data for this year
            image_props.append(np.nan)
    return np.array(html_props), np.array(image_props)

ia_html, ia_image = prep_props(ia)
kb_html, kb_image = prep_props(kb)

x = np.arange(len(years_full))
width = 0.8  # single wide bar per archive (since IA and KB are on separate plots)

fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

# --- IA plot ---
ax = axes[0]
ax.bar(x, ia_html, width, label="HTML", color="#1f77b4")
ax.bar(x, ia_image, width, bottom=ia_html, label="IMAGE", color="#ff7f0e")
ax.set_ylabel("Proportion")
ax.set_title("Proportion of HTML and IMAGE per Year â Internet Archive (IA)")
ax.set_ylim(0, 1)
ax.grid(axis="y", linestyle="--", linewidth=0.5)
ax.legend()

# --- KB plot ---
ax = axes[1]
ax.bar(x, kb_html, width, label="HTML", color="#1f77b4")
ax.bar(x, kb_image, width, bottom=kb_html, label="IMAGE", color="#ff7f0e")
ax.set_ylabel("Proportion")
ax.set_title("Proportion of HTML and IMAGE per Year â KB")
ax.set_ylim(0, 1)
ax.grid(axis="y", linestyle="--", linewidth=0.5)
ax.legend()

# Shared x-axis
plt.xticks(x, years_full, rotation=45)
axes[1].set_xlabel("Year")

plt.tight_layout()
plt.show()
```

The proportional distribution of HTML and IMAGE formats across the two web collections shows a clear long-term shift in the way the Rijksmuseum website has been captured and preserved. In the early years, particularly in the Internet Archive, image files constitute a substantial proportion of the archived resources. Over time, however, both archives show a pronounced move toward an HTML-dominated capture, reflecting changes in the captured museumâs web infrastructure. In general, this trend reflects a transition in the construction of the Rijksmuseumâs website to dynamically generated architecture, where images and other media are getting delivered from the external services (first of all related to JavaScript) that are harder for crawlers to capture.

As a result, the archival record by today is getting shaped fully based on the structural and textual components of the website and almost fully omits its visual elements. This means that the materials captured by both web collections are increasingly imbalanced in representation of the museumâs online presence and expose an example of a conflict between contemporary web technologies and crawler capabilities. HTML pages remain retrievable to a certain extent and dominate in the preserved material, whereas high-resolution and other types of images, dynamically loaded through external platforms and services, but are meaningful as part of the born digital heritage, are captured only sporadically or not at all.

#### Segmentation

The URL segmentation was implemented to identify the websiteâs meaningful substructures through content analysis of these segments. HTML type of files were selected for segmentation because they represent a foundational structure of the website.

During segmentation, the â/â sign has been used as a segment identifier. All the fragments of strings were separated according to the hierarchy and marked by the number of the segment. The number was assigned according to the order, following the domain name (see an example below).

```python
from IPython.display import HTML

url = "https://www.rijksmuseum.nl/nl/ontdek-de-collectie/overzicht/johannes-vermeer"

# Domain we want to segment after
prefix = "https://www.rijksmuseum.nl"

# Split URL into before-domain and after-domain parts
if prefix in url:
    before, after = url.split(prefix, 1)
else:
    before = url
    after = ""

# Split segments after the domain (ignore empty strings)
segments = [seg for seg in after.split('/') if seg]

# Build structured HTML for each segment with numbering
segment_boxes = ""
for i, seg in enumerate(segments, start=1):
    segment_boxes += f"""
        <span style="display: inline-block; text-align: center; margin-right: 5px;">
            <div style="font-size: 12px; color: #555;">{i}</div>
            <div style="
                text-decoration: underline;
                text-decoration-color: red;
                text-decoration-thickness: 4px;
                text-underline-offset: 4px;
                font-weight: bold;
            ">
                {seg}
            </div>
        </span>
        /
    """

# Combine full URL with domain and formatted segments
highlighted_url = f"{before}{prefix}/" + segment_boxes.rstrip("/")

# Reference text
reference_note = """
<div style="font-size: 13px; color: #555; margin-top: 8px;">
    This URL has been taken from the Internet Archive collection. It has records of archiving between 2013 and 2016.
</div>
"""

# Display inside a framed box (NO hyperlink)
html = f"""
<div style="
    border: 2px solid #4A90E2;
    padding: 12px 16px;
    margin: 12px 0;
    background-color: #F2F8FF;
    display: inline-block;
    border-radius: 6px;
    font-size: 16px;
    color: #003366;
">
    <b>{highlighted_url}</b>
    {reference_note}
</div>
"""

HTML(html)
```

In the URLs, there are several main types of segments:
1) prefix;
2) core parts of the website;
3) content-specific identifiers;
4) additional parameters and identifiers in the hierarchy.


A prefix demonstrates the language versioning (such as ânlâ for Dutch, âenâ for English, âukâ first used for British English but then assigned to Ukrainian, âdeâ for German, âfrâ for French). On the diagram below, there is a distribution of web collection by prefixes. From the archived URLs we see that the approach of the museum to publishing content in different languages varied. The pages in the Dutch and English are dominant, but the proportions of the prefixes are different for the Internet Archive and the KB Web Collection.

The comparison of language prefixes across the two web collections shows different preservation profiles shaped by the scope and harvesting strategies of the analysed institutions. The Internet Archive demonstrates more multilingual content on the Rijksmuseum website and shows that early captures from 2000 to 2004 were considerably more diverse in versioning, with English representing up to half of all accessible pages and several additional languages, including German, French, Spanish, and Italian.
```python
# Wordcloud without prefixes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import ipywidgets as widgets
from IPython.display import display, clear_output

# === 1. Load the resulting TOP15 table ===
csv_path = "./media/IA_url_segment_counts_by_year_TOP15_no_prefixes.csv"

df = pd.read_csv(csv_path, dtype=str)

# Identify segment columns and year columns
segment_cols = [c for c in df.columns if c.startswith("Segment")]
year_cols = [c for c in df.columns if c.isdigit()]
year_cols = sorted(year_cols, key=int)

# Convert numeric columns (years + total)
numeric_cols = year_cols + ['total'] if 'total' in df.columns else year_cols
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce").fillna(0)

# === helper: build word frequencies from a weight Series (per row) ===
def build_word_freq(weights: pd.Series):
    freq = {}
    for idx, w in weights.items():
        if w <= 0:
            continue
        row = df.loc[idx]
        for col in segment_cols:
            if col not in df.columns:
                continue
            word = row[col]
            if pd.isna(word):
                continue
            word = str(word).strip()
            if not word:
                continue
            freq[word] = freq.get(word, 0) + w
    return freq

# === helper: plot wordcloud with *mitigated* size differences ===
def plot_wordcloud(word_freq_raw, title):
    if not word_freq_raw:
        print("No data for this selection.")
        return

    values = np.array(list(word_freq_raw.values()))

    # Soft size scaling
    log_vals = np.log1p(values)
    scaled = (log_vals - log_vals.min()) / (log_vals.max() - log_vals.min() + 1e-9)
    scaled = 1 + scaled  # all sizes in [1, 2]

    word_freq = dict(zip(word_freq_raw.keys(), scaled))

    wc = WordCloud(
        width=1200,
        height=600,
        background_color="white"
    ).generate_from_frequencies(word_freq)

    plt.figure(figsize=(14, 7))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(title, fontsize=16, pad=40)   # â adds more space below the title
    plt.tight_layout()
    plt.show()

# === Interactive Layout ===

min_year = int(min(year_cols))
max_year = int(max(year_cols))

output = widgets.Output()

slider = widgets.IntRangeSlider(
    min=min_year,
    max=max_year,
    value=[min_year, max_year],
    step=1,
    description='Years',
    layout=widgets.Layout(width='80%', margin='20px 0px 0px 0px')  # â pushes slider DOWN
)

def update_wordcloud(change=None):
    start_year, end_year = slider.value
    years = [str(y) for y in range(start_year, end_year + 1)]
    weights = df[years].sum(axis=1)
    wf_raw = build_word_freq(weights)

    with output:
        clear_output(wait=True)
        plot_wordcloud(
            wf_raw,
            f"Wordcloud of top 15 segments collected per year {start_year}â{end_year}, Internet Archive"
        )

slider.observe(update_wordcloud, names='value')

# Initial display
update_wordcloud()

display(widgets.VBox([output, slider]))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Optional: for the Venn diagram (if installed)
try:
    from matplotlib_venn import venn2
    HAS_VENN = True
except ImportError:
    HAS_VENN = False
    print("matplotlib_venn not installed â Venn diagram will be skipped.")

# --- File paths ---
kb_path = "./media/KB_Object_Numbers.csv"
ia_path = "./media/IA_Object_Numbers.csv"

# --- Read files ---
kb = pd.read_csv(kb_path)
ia = pd.read_csv(ia_path)

# --- Get unique object numbers as sets ---
kb_set = set(kb['object_number'].dropna().astype(str).str.strip())
ia_set = set(ia['object_number'].dropna().astype(str).str.strip())

# --- Overlap & uniques ---
overlap = ia_set & kb_set
ia_only = ia_set - kb_set
kb_only = kb_set - ia_set

n_ia_only = len(ia_only)
n_kb_only = len(kb_only)
n_overlap = len(overlap)

print("=== OBJECT NUMBER OVERLAP SUMMARY ===")
print(f"KB unique object_numbers: {len(kb_set)}")
print(f"IA unique object_numbers: {len(ia_set)}")
print(f"Overlap (in both):        {n_overlap}")
print(f"Only in KB:               {n_kb_only}")
print(f"Only in IA:               {n_ia_only}")

# --- VENN / SPHERIC DIAGRAM ONLY ---
if HAS_VENN:
    plt.figure(figsize=(6, 6))
    v = venn2(
        subsets=(n_kb_only, n_ia_only, n_overlap),
        set_labels=('KB', 'IA')
    )

    # Colours
    v.get_patch_by_id('10').set_color('#ff7f0e')  # KB only
    v.get_patch_by_id('01').set_color('#1f77b4')  # IA only
    v.get_patch_by_id('11').set_color('#6a3d9a')  # overlap

    # Move KB-only number outside the left circle
    kb_only_label = v.get_label_by_id('10')
    if kb_only_label is not None:
        x, y = kb_only_label.get_position()
        kb_only_label.set_position((x - 0.06, y))  # gentle shift left

    # Reposition KB label upward
    kb_set_label = v.get_label_by_id('A')
    if kb_set_label is not None:
        x, y = kb_set_label.get_position()
        kb_set_label.set_position((x - 0.08, y + 0.05))

    plt.title("Overlap of object_number between KB and IA")
    plt.show()
```
